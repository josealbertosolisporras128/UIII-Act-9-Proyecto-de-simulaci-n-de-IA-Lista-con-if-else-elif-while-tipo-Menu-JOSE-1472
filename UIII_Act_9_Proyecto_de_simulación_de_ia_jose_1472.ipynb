```python
# Inicialización: La lista de servicios empieza vacía [1, 2]
carrito_estilo = []
presupuesto = 0.0

print("--- Bienvenido a StyleBot: Tu Consultor de Imagen con IA ---")
presupuesto = float(input("Para empezar, ¿cuál es tu presupuesto hoy?: "))

# Bucle infinito para mantener el menú interactivo [3, 4]
while True:
    print("\n--- MENÚ DE ASESORÍA IA ---")
    print("1. IA: Recomendar Corte de Cabello ($30)")
    print("2. IA: Recomendar Tinte de Color ($70)")
    print("3. Ver mis servicios elegidos (Lista)")
    print("4. Vaciar lista de deseos")
    print("5. Salir")

    opcion = input("\nSelecciona una opción (1-5): ")

    # Opción 1: Lógica de recomendación de corte
    if opcion == "1":
        if presupuesto >= 30: # Operador relacional [5, 6]
            print("\nIA: Analizando forma de rostro...")
            rostro = input("¿Tu rostro es 'Redondo' u 'Ovalado'?: ")

            # Operador lógico 'or' para ser flexible con mayúsculas [7, 8]
            if rostro == "Redondo" or rostro == "redondo":
                sugerencia = "Corte Pixie con volumen"
                print(f"IA sugiere: {sugerencia} para alargar tu perfil.")
            else:
                sugerencia = "Corte Bob en capas"
                print(f"IA sugiere: {sugerencia} para resaltar tu simetría.")

            # Agregar elemento a la lista mediante el método append() [1, 9]
            carrito_estilo.append(sugerencia)
            presupuesto = presupuesto - 30 # Operador aritmético [6, 10]
            print(f"¡'{sugerencia}' añadido a tu lista! Saldo restante: ${presupuesto}")
        else:
            print("IA: Error. No tienes presupuesto suficiente para un corte.")

    # Opción 2: Lógica de tinte
    elif opcion == "2":
        if presupuesto >= 70:
            print("\nIA: Escaneando tono de piel...")
            tono = input("¿Tu piel es de tono 'Cálido' o 'Frío'?: ")

            if tono == "Cálido" or tono == "cálido":
                color = "Tinte Cobrizo Dorado"
            elif tono == "Frío" or tono == "frío":
                color = "Tinte Cenizo Platino"
            else:
                color = "Tinte Castaño Neutro"

            print(f"IA recomienda: {color}")
            carrito_estilo.append(color) # Guardar en la lista [2, 11]
            presupuesto = presupuesto - 70
            print(f"¡'{color}' añadido! Saldo restante: ${presupuesto}")
        else:
            print("IA: Saldo insuficiente para procesos de color.")

    # Opción 3: Mostrar elementos de la lista sin usar ciclo 'for'
    elif opcion == "3":
        print("\n--- TUS SERVICIOS ELEGIDOS ---")
        if len(carrito_estilo) == 0: # Verificar si la lista está vacía [12, 13]
            print("Tu lista de deseos está vacía actualmente.")
        else:
            # Usamos un ciclo while y un contador para recorrer la lista [4, 14]
            indice = 0
            while indice < len(carrito_estilo):
                print(f"- {carrito_estilo[indice]}") # Acceso por índice [1, 2]
                indice = indice + 1
            print(f"Presupuesto restante: ${presupuesto}")

    # Opción 4: Vaciar la lista
    elif opcion == "4":
        carrito_estilo = [] # Reiniciar a lista vacía [1, 2]
        print("IA: Tu lista de servicios ha sido borrada.")

    # Opción 5: Salir del sistema
    elif opcion == "5":
        print("IA: Gracias por usar StyleBot. ¡Disfruta tu nueva imagen!")
        break # Rompe el bucle while True [15, 16]

    else:
        print("IA Error: Opción no válida. Por favor, elige del 1 al 5.")
```
